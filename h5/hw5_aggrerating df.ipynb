Load dataset nyc_weather_2018 and weather_stations, then calculate the average elevation of stations that calculat‘SNOW’ and contains ‘NJ US’ in its name.

```python
import pandas as pd
import numpy as np
```

```python
station_info = pd.read_csv("weather_stations.csv")
nyc_weather = pd.read_csv("nyc_weather_2018.csv")
```

```python
station_info.head()
```

```python
nyc_weather.head()
```

```python
station_info = station_info.rename(columns={'id':'station'})
station_info.head()
```

```python
dt = station_info.merge(nyc_weather, on = 'station')
dt.head()
```

```python
avg = dt.query('datatype== "SNOW" and name.str.contains("NJ US")', engine='python').groupby('station').mean()
avg.elevation.mean()
```

Load the fb dataset, then:
•Calculate the percentage change between that and and last day in volumn (pct_change method)
•Sort the dataframe by percentage change then get five days that have the biggest change

```python
fb_2018 = pd.read_csv("fb_2018.csv")
fb_2018.head()
```

```python
fb_2018 = fb_2018.assign(pct_change = lambda x: x.volume.pct_change())
print(fb_2018.head())
```

```python
fb_2018.sort_values(by='pct_change', ascending=True)
fb_2018.head(5)
```
